# Training trends

## Model 1: lambda = 1

```python
training_losses = np.array([ 6.007, 4.288,3.646,3.245  ])
val_losses = np.array([4.915, 4.042,3.626, 3.455])
epoch = np.array([1,2,3,4])
plt.plot(epoch, training_losses, label = "Training")
plt.plot(epoch, val_losses, label = "Validation")
plt.legend()
plt.title("Loss for the model with lambda = 1")
plt.xlabel("Epoch")
plt.ylabel("Loss")
```
